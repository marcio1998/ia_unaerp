```python
E1 = [0,0,1,1]
E2 = [0,1,0,1]
saida = [0.0,1.0,1.0,0.0]
tx = 0.0001
erro_inicial = 0.0
I0 = np.zeros(4)
O0 = np.zeros(4)
I1 = np.zeros(4)
O1 = np.zeros(4)
I2 = np.zeros(2)
O2 = np.zeros(2)
w1 = np.random.rand(4,4)
w2 = np.random.rand(4,2)
nw1 = np.zeros((4,4))
nw2 = np.zeros((4,2))
vw1 = np.zeros((4,4))
vw2 = np.zeros((4,2))
d2 = 0.0
d1 = np.zeros(3)
```

```python
def manipular_peso_trash_hold(matriz):
  if matriz.shape[0] == 2:
    for i in range(matriz.shape[0]):
      for j in range(matriz.shape[1]):
        matriz[i][j] =  (matriz[i][j]*40) -20 
  else:
    for i in range(matriz.shape[0]):
      matriz[i] = (matriz[i]*40)-20
```
